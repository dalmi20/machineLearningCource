```python
import pandas as pd
df=pd.read_csv("salaries.csv")
df.head()
```

```python
#independent variable
inputs = df.drop('salary_more_then_100k',axis='columns')
#dependent variable(target)
target=df['salary_more_then_100k']
```

```python
inputs
```

```python
target
```

```python
#transform the categorical variables to numerics one 
from sklearn.preprocessing import LabelEncoder
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
#create new columns with numeric values
inputs['company_n']=le_company.fit_transform(inputs['company'])
inputs['job_n']=le_company.fit_transform(inputs['job'])
inputs['degree_n']=le_company.fit_transform(inputs['degree'])
inputs.head()
```

```python
inputs_n=inputs.drop(['company','job','degree'],axis='columns')
inputs_n.head()
```

```python
#google=2 abcPharma=0 facebook =1 same thing with job and degree
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(inputs_n,target,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[2,0,1]])
```

```python
tit=pd.read_csv('titanic.csv')
tit.head()
```

```python
tit_n=tit.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived'],axis='columns')
tit_n.head()
```

```python
survived=tit['Survived']
survived
```

```python
tit_n.Sex = tit_n.Sex.map({'male': 1, 'female': 2})
merged_n=tit_n
merged_n
```

```python
medianAge=merged_n.Age.median()
medianAge
```

```python
medianPclass=merged_n['Pclass'].median()
medianPclass
```

```python
medianFare=merged_n['Fare'].median()
medianFare
```

```python
merged_n.Age=merged_n.Age.fillna(medianAge)
merged_n.Pclass=merged_n.Pclass.fillna(medianPclass)
merged_n.Fare=merged_n.Fare.fillna(medianFare)
merged_n
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(merged_n,survived,test_size=0.2)
```

```python
x1_train
```

```python
x1_test
```

```python
y1_train
```

```python
model.fit(x1_train,y1_train)
```

```python
model.predict(x1_test)
```

```python
model.score(x1_test,y1_test)
```
