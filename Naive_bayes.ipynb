```python
import pandas as pd
df= pd.read_csv("titanic.csv")
df.head()
```

data Exploration 

```python
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis='columns',inplace=True)
df.head()
```

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

```python
dummies=pd.get_dummies(inputs.Sex) # give a true false values instead of 0 1 soo 
dummies=dummies.astype(int)
dummies
```

```python
merged=pd.concat([inputs,dummies],axis='columns')
merged.head()
```

```python
merged=merged.drop("Sex",axis='columns')
merged
```

```python
#how we found missing values in each column
merged.columns[merged.isna().any()]
```

```python
merged.Age[:10]
```

```python
merged.Age=merged.Age.fillna(merged.Age.mean())
merged.head(10)
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test=train_test_split(merged,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
#we use GaussianNB() when the distribution is normal 
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_test[:10]
```

```python
model.predict(x_test[:10])
```

Email spam 

```python
import pandas as pd 
df=pd.read_csv("spam.csv")
df.head()
```

```python
df.groupby("Category").describe()
```

```python
df['spam']=df["Category"].apply(lambda x:1 if x =='spam' else 0)
df.head()
dff=df.drop('Category',axis="columns")
dff
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dff.Message,dff.spam, test_size=0.25)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
v=CountVectorizer()
x_train_count=v.fit_transform(x_train.values)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails=[
    'hey ferdaous, i hope you doing well',
    "URGENT! You have won a 1 week FREE membership in our £100,000 Prize Jackpot! Txt the word: CLAIM to No: 81010 T&C www.dbuk.net LCCLTD POBOX 4403LDNW1A7RW18"
]
emails_count=v.transform(emails)
model.predict(emails_count)
```

```python
x_test_count=v.transform(x_test)
model.score(x_test_count,y_test)
```

```python
from sklearn.pipeline import Pipeline
clf=Pipeline([('vectorize',CountVectorizer()),('nb',MultinomialNB())]) 
```

```python
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
clf.predict(emails)
```
