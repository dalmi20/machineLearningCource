Identify hand written digits recognition

```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
dir(digits)
```

```python
digits.data[0]
```

```python
plt.gray()
for i in range(5): 
    plt.matshow(digits.images[i])
```

```python
digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(digits.data,digits.target,test_size=0.2)
#independent variables = digits.data
#dependent variables = digits.target
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
plt.matshow(digits.images[78])
```

```python
digits.target[78]
```

```python
model.predict([digits.data[78]])
```

```python
model.predict(digits.data[0:5])
```

```python
# the score of this model is 97% we want to know where the model failed
#soo we using confusion matrix
y_predicted= model.predict(X_test)
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,y_predicted)
cm
#in this matrix if the score = 1 soo  we will find 0 in the entire matrix except  diagonal 
#else we will find how many the model fail in prediction 
```

```python
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
